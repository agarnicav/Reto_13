```python
import json


def leer_json(nombre_archivo):
    with open(nombre_archivo, 'r') as archivo:
        contenido = archivo.read()
        datos = json.loads(contenido)
        return datos

def buscar_por_deporte(datos, deporte):
    for  Dato in datos:
        for persona, informacion in Dato.items():
            if deporte in informacion['deportes']:
                nombres = informacion['nombres']
                apellidos = informacion['apellidos']
                print("Nombres completos:" +str(nombres)+  "  " +str(apellidos))

def buscar_por_rango_edad(datos, edad_min, edad_max):
    for Dato in datos:
        for persona, informacion in Dato.items():
            edad = informacion['edad']
            if edad >= edad_min and edad <= edad_max:
                nombres = informacion['nombres']
                apellidos = informacion['apellidos']
                print("Nombres completos:" +str(nombres)+ " "+ str(apellidos))
                
                

if __name__ == "__main__":

    # Ruta y nombre de archivo JSON
    nombre_archivo = "datos.json"

    # Leer el archivo JSON
    datos = leer_json(nombre_archivo)

    # Búsqueda por deporte
    deporte = input("Ingrese el deporte a buscar: ")
    print("las Personas o persona que practican el deporte " + str(deporte)+":")
    buscar_por_deporte(datos, deporte)

    # Búsqueda por rango de edad
    edad_min = int(input("Ingrese la edad mínima: "))
    edad_max = int(input("Ingrese la edad máxima: "))
    print("Personas dentro del rango de edades de : "+str(edad_min)+"-" +str(edad_max))
    buscar_por_rango_edad(datos, edad_min, edad_max)
```
